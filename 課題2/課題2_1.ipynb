```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import os
```

```python
def scrape():
  url="https://www.saitama-u.ac.jp/"
  res = requests.get(url)
  soup = BeautifulSoup(res.text, 'html.parser')
  news = soup.find_all("li",class_="mtx line news")
  content=[]
  p1=r"\[.+\]"
  p2="\[|\]"
  for n in news:
    content.append(
        [re.sub(p2,'',n.find('span').next_element),re.sub(p1, '', n.get_text())]
    )
  return content
```

```python
df_n=pd.DataFrame(scrape(),columns=["date","contents"])
file="./saitama_u_news.csv"
if os.path.isfile("./saitama_u_news.csv"):
  df_b=pd.read_csv(file)
  df=pd.concat(
      [df_n,df_b],
      axis=0,
      ignore_index=True
  )
else:
  df=df_n
df.drop_duplicates()
df.to_csv(file,index=False,encoding='utf-8')
```
